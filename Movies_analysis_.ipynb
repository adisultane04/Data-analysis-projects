# Data Visualization

```python
df['Genre'].value_counts().plot(kind='bar')
plt.title("Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
df
```

```python
df.groupby('Genre')['IMDB_Rating'].mean().plot()
plt.title("Average Amount Over Time")
plt.xlabel("Date")
plt.ylabel("Amount")
plt.show()
```

```python
df['IMDB_Rating'].plot(kind='hist', bins=10)
plt.title("Amount Distribution")
plt.show()
```
